## 2. Create a histogram for each feature in the dataset to illustrate the feature distributions

```python
import seaborn as sns 
import matplotlib.pyplot as plt
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(ncols=2,nrows=2,figsize=(12,12),sharey=True)

sns.histplot(iris['SepalLengthCm'],ax=ax1,linewidth=0.3,color='#77dd33')
sns.histplot(iris['SepalWidthCm'],ax=ax2,linewidth=0.3,color='#77dd33')
sns.histplot(iris['PetalLengthCm'],ax=ax3,linewidth=0.3,color='#77dd33')
sns.histplot(iris['PetalWidthCm'],ax=ax4,linewidth=0.3,color='#77dd33')
```

## 3. Create a boxplot for each feature in the dataset.

```python
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(ncols=2,nrows=2,figsize=(12,12))

sns.boxplot(iris['SepalLengthCm'],ax=ax1,linewidth=1,color='#55bb55')
sns.boxplot(iris['SepalWidthCm'],ax=ax2,linewidth=1,color='#55bb55')
sns.boxplot(iris['PetalLengthCm'],ax=ax3,linewidth=1,color='#55bb55')
sns.boxplot(iris['PetalWidthCm'],ax=ax4,linewidth=1,color='#55bb55')
```

## 4. Compare distributions and identify outliers.

```python
measures = list(iris.columns)
measures.remove('Id')
measures.remove('Species')
measures
```

```python
def get_outliers(df):
    z_score = (df - df.mean())/ df.std()
    return df[(abs(z_score) >= 2.5).any(axis=1)]
```

```python
get_outliers(iris[measures])
```
