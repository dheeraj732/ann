## 1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can find any patterns in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
fig,ax = plt.subplots(figsize=(12,5))
sns.scatterplot(y=titanic['Fare'],x=np.arange(titanic.shape[0]),hue=titanic['Survived'],palette='cool',ax=ax,linewidth=0)

# titanic.groupby('Survived')['Age'].mean()
```

```python
x1 = titanic[titanic['Survived']==1]['Age']
x2 = titanic[titanic['Survived']==0]['Age']
fig,(ax1,ax2) = plt.subplots(figsize=(12,5),ncols=2,sharey=True)
ax1.grid(axis='y',alpha=0.2)
ax2.grid(axis='y',alpha=0.2)
sns.histplot(x1,color='#9988ff',alpha=0.7,ax=ax1,label='Survived',linewidth=0.4);
sns.histplot(x2,color='#33ff22',alpha=0.9,ax=ax2,label='Not Survived',linewidth=0.4);
fig.legend();
```

```python
x1 = titanic[titanic['Survived']==1]['Pclass'].value_counts()
x2 = titanic[titanic['Survived']==0]['Pclass'].value_counts()
with plt.style.context('default'):
    fig,ax = plt.subplots()
    ax.grid(axis='y',alpha=0.2)
    sns.barplot(x2,color='blue',alpha=0.5,label='not survived',ax=ax)
    sns.barplot(x1,color='green',alpha=0.5,label='survived',ax=ax)
    ax.legend();
```

```python
x1 = titanic[titanic['Survived']==1]['Parch'].value_counts()
x2 = titanic[titanic['Survived']==0]['Parch'].value_counts()
fig,ax = plt.subplots()
ax.grid(axis='y')
sns.barplot(x2,color='blue',alpha=0.8,label='not survived',ax=ax)
sns.barplot(x1,color='green',alpha=0.8,label='survived',ax=ax)
ax.legend();
```

```python
x1 = titanic[titanic['Survived']==1]['Embarked'].value_counts()
x2 = titanic[titanic['Survived']==0]['Embarked'].value_counts()
fig,ax = plt.subplots()
ax.grid(axis='y')
sns.barplot(x2,color='blue',alpha=0.5,label='not survived',ax=ax)
sns.barplot(x1,color='green',alpha=0.5,label='survived',ax=ax)
ax.legend();
```
